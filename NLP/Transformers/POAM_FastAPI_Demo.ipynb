## Build out full methods for NLP implimentation 

```python
import scispacy
import spacy
nlp = spacy.load("en_ner_bc5cdr_md") 

import stanza
stanza.download('en')
en_nlp = stanza.Pipeline('en',verbose=False)

nlp2 = stanza.Pipeline(lang='en', processors='tokenize,ner')

stanza.download('en', package='mimic', processors={'ner': 'i2b2'})
nlp3 = stanza.Pipeline('en', package='mimic', processors={'ner': 'i2b2'})

from transformers import pipeline
classifier = pipeline("sentiment-analysis")
```

```python

def extract_medical_entitiys(text: str):
    doc = nlp(text)
    res = []
    #print("TEXT", "START", "END", "ENTITY TYPE")
    for ent in doc.ents:
      sets = {}
      #print(ent.text, ent.start_char, ent.end_char, ent.label_)
      sets['TEXT'] = ent.text
      sets['START'] = ent.start_char
      sets['END'] = ent.end_char
      sets['ENTITY TYPE'] = ent.label_
      res.append(sets)

    return(res)


def extract_sentence_structure(text: str):
    en_doc = en_nlp(text)
    res = []
    for i, sent in enumerate(en_doc.sentences):
        #print("[Sentence {}]".format(i+1))
        for word in sent.words:
          #print("{:12s}\t{:12s}\t{:6s}\t{:d}\t{:12s}".format(word.text, word.lemma, word.pos, word.head, word.deprel))
          #print("")
          sets = {}
          sets['Text'] = word.text
          sets['Lemma'] = word.lemma
          sets['Part-of-Speech'] = word.pos
          sets['head'] = word.head
          sets['Deprel'] = word.deprel
          res.append(sets)

    return res


def extract_sentence_ner(text: str):
    doc = nlp2(text)
    res = []
    #print()
    #print(*[f'entity: {ent.text}\ttype: {ent.type}' for ent in doc.ents], sep='\n')
    for ent in doc.ents:
      sets = {}
      sets['Text']= ent.text
      sets['Type']= ent.type
      res.append(sets)
    
    return res

def clasifiy_medical_text(text: str):
    doc = nlp3(text)
    res = []

    # print out the entities
    for ent in doc.entities:
        #print(f'{ent.text}\t{ent.type}') 
        sets = {}
        sets['Text']= ent.text
        sets['Type']= ent.type
        res.append(sets)

    return res


def clasify_outcome(text: str):
    result = classifier(text)[0]
    #print(f"label: {result['label']}, with score: {round(result['score'], 4)}")
    res = f"label: {result['label']}, with score: {round(result['score'], 4)}"
    return res



#run tests - 
test_text ='The patient had a dry cough and fever, they were treated with Paracetamol and saw major improvements in the next few weeks'

#test 1 - extract medical entitys
#extract_medical_entitiys(test_text)


#test 2 - extract sentence structure 
#extract_sentence_structure(test_text)


# test 3 - extract sentence ner 
#extract_sentence_ner(test_text)


#test 4 clasified medical text
#clasifiy_medical_text(test_text)


#texst 5 get clasification of 
#clasify_outcome(test_text)
```

# Using Fast API

Let’s now quickly jump to FastAPI which is a very promising library of Python due to the fact that it is based on ASGI and handles requests asynchronously increasing throughput and performance as compared to Flask which uses WSGI that doesn’t support async calls. FastAPI has a good documentation available out there as well. In order to run FastAPI on Google Colab, we need to install it first along with the following packages:

https://fastapi.tiangolo.com/tutorial/request-files/ 

```python
from fastapi import FastAPI, File, UploadFile
import nest_asyncio
from pyngrok import ngrok
import uvicorn

app = FastAPI()

@app.get('/index')
async def home():
  return "Test Success : API status Healthy"


@app.get('/patient_outcome')
async def outcome_s(text :str= 'The patient had a dry cough and fever, they were treated with Paracetamol and saw major improvements in the next few weeks'):
  res = clasify_outcome(text)
  return {'inital text':text, 'result':res}


@app.get('/extract_medical')
async def medical_e(text :str= 'The patient had a dry cough and fever, they were treated with Paracetamol and saw major improvements in the next few weeks'):
  res = extract_medical_entitiys(text)
  return {'inital text':text, 'result':res}


@app.get('/extract_ner')
async def ner_e(text :str= 'The patient had a dry cough and fever, they were treated with Paracetamol and saw major improvements in the next few weeks'):
  res = extract_sentence_ner(text)
  return {'inital text':text, 'result':res}


@app.get('/extract_structure')
async def structure_e(text :str= 'The patient had a dry cough and fever, they were treated with Paracetamol and saw major improvements in the next few weeks'):
  res = extract_sentence_structure(text)
  return {'inital text':text, 'result':res}


@app.get('/extract_&_classify_medical_text')
async def classify_medical_e(text :str= 'The patient had a dry cough and fever, they were treated with Paracetamol and saw major improvements in the next few weeks'):
  res = clasifiy_medical_text(text)
  return {'inital text':text, 'result':res}




ngrok_tunnel = ngrok.connect(8000)
print('Public URL:', ngrok_tunnel.public_url)
nest_asyncio.apply()
uvicorn.run(app, port=8000)
```
